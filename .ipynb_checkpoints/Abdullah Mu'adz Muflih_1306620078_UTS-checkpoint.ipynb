## Time Domain

#### Energy

```python
def energy(x,sr):
    hop_length = 256
    frame_length = 512
    energi = np.array([
    sum(abs(x[i:i+frame_length]**2))
    for i in range(0, len(x), hop_length)])
    plt.figure(figsize=(18,10))
    t = librosa.get_duration(x, sr=sr, hop_length=hop_length)
    x = np.linspace(0,t,len(energi))
    plt.plot(x,energi, ls='-', c = 'r')
    plt.xlabel('waktu')
    plt.ylabel('energi')
    plt.show()
    return x,energi

waktu,energi = energy(audio,sr_au)
```

```python
counts, bins = np.histogram(energi)
mean = statistics.mean(energi)
sd = statistics.stdev(energi)
x = mean + sd * np.random.randn(437)

y = ((1 / (np.sqrt(2 * np.pi) * sd)) *
     np.exp(-0.5 * (1 / sd * (bins - mean))**2))

plt.figure(figsize=(18,10))
plt.hist(bins[:-1], bins, weights=counts)
plt.title('Energy - Histogram')
plt.xlabel('Energi')
plt.ylabel('Jumlah')

plt.figure(figsize=(18,10))
plt.plot(bins, y, '--',c='k')
plt.title('Energy - stdbymean')
plt.xlabel('Energi')
plt.ylabel('Jumlah')
plt.show()
```

###### Analisis Energi
- Instrumen Tiup
    - Plot energi terlihat lebih smooth saat instrumen dibunyikan
    - Energi pada tiupan pertama naik dari note yang ditiup pertama ke yang ke dua
    - Tiupan pertama memiliki rata - rata energi yang lebih kecil dibandingkan instrumen kedua.
    - Energi pada tiupan kedua, note kedua mengalami kenaikan dan landaian energi.
    
    
    
- Suara
    - Energi yang dihasilkan sangat noisy
    - Terdapat 3 peak energi pada suara sesi pertama dan dua peak energi pada suara sesi kedua

#### Zero-Crossing rate

```python
hop_length = 256
frame_length = 2048 #semakin besar semakin jelas garisnya

ZC = librosa.feature.zero_crossing_rate(y = audio, frame_length=frame_length, hop_length=hop_length)
plt.figure(figsize=(18,10))
librosa.display.waveshow(audio, sr=sr_au)

plt.figure(figsize=(18,10))
plt.plot(waktu,ZC[0])
plt.ylim(-(max(ZC[0])))

plt.show()
```

```python
counts, bins = np.histogram(ZC[0])
mean = statistics.mean(ZC[0])
sd = statistics.stdev(ZC[0])
x = mean + sd * np.random.randn(437)

y = ((1 / (np.sqrt(2 * np.pi) * sd)) *
     np.exp(-0.5 * (1 / sd * (bins - mean))**2))

plt.figure(figsize=(18,10))
plt.hist(bins[:-1], bins, weights=counts)
plt.title('Energy - Histogram')
plt.xlabel('Zero Crossed')
plt.ylabel('Jumlah')
plt.show()
```

###### Zero-Crossing Rate
- Zero-Crossing rate adalah seberapa banyak audio melakukan perubahan antara value positif dan negatif.


- Instrumen Tiup
    - Tiupan pertama
        - Terdapat perubahan dari ZC Rate tinggi ke rendah, kemungkinan akibat perubahan note.
    - Tiupan kedua
        - Terlihat datar saja tetapi terlihat 3 perbedaan, di dataran pertama crosing ratenya berbentuk segitiga (ZC Rate berubah-ubah), lalu bentuk dataran dan ada lembah, dan terakhir terdapat turunan, perbedaan ini kemungkinan karena perbedaan note.
    
    
- Suara
    - Omongan pertama
        - Terdapat 3 gunung ZC Rate yang tingginya mirip.
    - Omongan kedua
        - Terdapat 3 gunung ZC Rate dengan gunung terakhir memiliki ZC Rate lebih tinggi dari 2 gunung sebelumnya.

## Frequency Domain

#### Spectral Centroid

```python
scent = librosa.feature.spectral_centroid(y=audio, sr=sr_au)

plt.figure(figsize=(18,10))
plt.semilogy(scent[0], label='Spectral centroid')
plt.ylabel('Hz')
plt.xlabel('')
plt.xticks([])
plt.xlim([0, scent.shape[-1]])
plt.legend()
plt.show()
```

###### Spectral Centroid

- Merupakan center of gravity pada spectrum (dapat dilihat pada grafik Mel Spectogram)


- Instrumen Tiup
    - Tiupan pertama
        - Memiliki frekuensi yang tinggi dan terjadi penurunan frekuensi
    - Tiupan kedua
        - Memiliki 3 perbedaan frekuensi, pertama tinggi, lalu turun dan rata, dan ketiga naik lagi tetapi tidak setinggi dan serta yang pertama, terdapat banyak perubahan frekuensi pada puncak ketika
    
    
- Suara
    - Omongan pertama
        - Terdapat 2 peak frekuensi peak kedua lebih tinggi dari peak pertama
    
    - Omongan kedua
        - Terdapat 3 peak frekuensi dan memiliki frekuensi yang mirip antara peak

#### Plot Koefisien Fourier

```python
n = audio.size

yf = fft(audio)
xf = fftfreq(n, 1/sr_au)

dat1 = {'freq':xf,'koef fourier':yf}
df1 = pd.DataFrame(dat1)
df1.to_csv(r'Koef_fourier_gitar1', index=False)

plt.figure(figsize=(20,12))
#Plot pertama tanpa pembesaran
plt.subplot(121)
plt.plot(xf/2, np.abs(yf))
plt.title('Plot FFT Audio 0 - 20.000 Hz')
plt.xlabel('Frequency (Hz)');
plt.ylabel('Power Spectral Density / Power (dB)')

#plot kedua dengan pembesaran
plt.subplot(122)
plt.plot(xf/2, np.abs(yf))
plt.xlim(-10, 2200)
plt.title('Plot FFT Audio -10 sd 2200 Hz')
plt.xlabel('Frequency (Hz)');
plt.ylabel('Power Spectral Density / Power (dB)')
plt.show()
print(n)
```

###### Transformasi Fourier

- Frekuensi dominan yang berbunyi terdapat pada frekuensi 0 - 400 Hz

#### MFCC

```python
mfcc = librosa.feature.mfcc(audio, sr=sr_au)
S1 = librosa.feature.melspectrogram(y=audio, sr=sr_au, n_mels=128,
                                   fmax=9000)

freqs, times, D = librosa.reassigned_spectrogram(audio, fill_nan=True)
```

```python
fig, ax = plt.subplots(nrows=2, sharex=True, figsize=(25,35))
img = librosa.display.specshow(librosa.power_to_db(S1, ref=np.max),
                               x_axis='time', y_axis='mel', fmax=8000,
                               ax=ax[0])
fig.colorbar(img, ax=[ax[0]])
ax[0].plot(times[:][-1], scent.T, label='Spectral centroid', 
           color='lightblue',linewidth=3)

ax[0].legend(loc='upper right')
ax[0].set(title='Mel spectrogram')
ax[0].label_outer()
img = librosa.display.specshow(mfcc, x_axis='time', ax=ax[1])
fig.colorbar(img, ax=[ax[1]])
ax[1].set(title='MFCC')
```
